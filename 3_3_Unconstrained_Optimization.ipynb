## Global vs. Local Minimizers

To understand the difference between global and local minimizers, we'll start by plotting a simple function.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 400)
y = x**4 - 4 * x**2 + 4

plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.title("Function: \(f(x) = x^4 - 4x^2 + 4\)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

From the plot, you can see two local minimizers and one local maximizer. However, only the one at the bottom-most part is the global minimizer.

```python
def gradient(x):
    return 4 * x**3 - 8 * x


def gradient_descent(lr, iterations):
    x = 2  # starting point
    history = [x]

    for _ in range(iterations):
        grad = gradient(x)
        x = x - lr * grad
        history.append(x)

    return history


history = gradient_descent(lr=0.01, iterations=1000)

plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.scatter(history, [i**4 - 4 * i**2 + 4 for i in history], color="red", s=10)
plt.title("Gradient Descent Convergence")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

The red points show the path of the gradient descent starting from x=2. As we can observe, the gradient descent method successfully approaches the global minimizer.

## Convex Functions and Their Properties

Convex functions have an essential property that any local minimizer is a global minimizer. We'll demonstrate this with a simple quadratic function.

```python
x = np.linspace(-2, 2, 400)
y = x**2

plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.title("Function: \(f(x) = x^2\)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

The function \(f(x) = x^2\) is convex, as shown above. It's evident that any local minimum of this function is also its global minimum.
