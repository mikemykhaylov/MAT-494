# 3.4 Logistic Regression

Logistic regression is a model that uses a logistic function to model a binary dependent variable. This demonstration will showcase the use of gradient descent for binary classification through logistic regression.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
```

## Dataset Generation
Let's generate a synthetic dataset to understand logistic regression better.

```python
# Generating a dataset
X, y = make_classification(
    n_samples=100,
    n_features=1,
    n_informative=1,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=42,
)

# Plotting the data
plt.figure()
plt.scatter(X, y, c=y, cmap="rainbow")
plt.title("Synthetic Data")
plt.xlabel("Feature Value")
plt.ylabel("Class")
plt.show()
```

## Logit Function and Sigmoid
The logit function is a graph that represents the logistic function. The sigmoid function, which is the logistic function, is given by:
$$\sigma(x) = \frac{1}{1 + e^{-x}}$$

```python
# Defining the sigmoid function
def sigmoid(t):
    return 1 / (1 + np.exp(-t))


# Generating values for t and their corresponding sigmoid values
t = np.linspace(-10, 10, 400)
s = sigmoid(t)

# Plotting the sigmoid function
plt.figure()
plt.plot(t, s)
plt.title("Sigmoid Function")
plt.xlabel("t")
plt.ylabel("σ(t)")
plt.grid(True)
plt.show()
```

## Logistic Regression with Gradient Descent

The objective is to minimize the cross-entropy loss:

$$l(x;A,b) = -\frac{1}{N}\sum_{i=1}^N y_i\log(\sigma(Ax_i + b)) + (1-y_i)\log(1-\sigma(Ax_i + b))$$

Let's use the sklearn implementation of Logistic Regression and fit the data.

```python
# Logistic Regression using sklearn
model = LogisticRegression()
model.fit(X, y)

# Predict probabilities
X_test = np.linspace(-3, 3, 300)
y_prob = model.predict_proba(X_test.reshape(-1, 1))[:, 1]

# Plotting the logistic regression curve
plt.figure()
plt.scatter(X, y, c=y, cmap="rainbow", label="Data")
plt.plot(X_test, y_prob, color="green", label="Logistic Regression Curve")
plt.title("Logistic Regression Fit")
plt.xlabel("Feature Value")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

## Gradient Descent Iteration
Using the gradient, the update rule for gradient descent is:

$$\theta_{t+1} = \theta_t - \alpha \nabla_\theta l(\theta_t)$$

where $\alpha$ is the learning rate. We won't manually implement gradient descent since we've used sklearn, but this equation depicts how weights are updated in each iteration.
